```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [32, 37, 0]
y_data = [89.6, 98.6, 32]
def forward(x):
    return x * w

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y)*(y_pred - y)

wlist = []; mselist = []

for w in np.arange(0.0, 4.0, 0.1):
    print("w = ", w)
    loss_sum = 0
    for x_val, y_val in zip(x_data, y_data):
        y_pred_value = forward(x_val)
        loss_val = loss(x_val, y_val)
        loss_sum += loss_val
        print("\t", x_val, y_val, y_pred_value, loss_val)
    print("MSE = ", loss_sum / 3)
    wlist.append(w)
    mselist.append(loss_sum / 3)
    
plt.plot(wlist, mselist)
plt.show()
```

```python
least_mse = print(min(mselist))
```
